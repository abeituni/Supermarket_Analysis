```python
#Import Dependencies & Load in Data

import pandas as pd
import numpy as np
import json

auditors = "../engage3project/auditors.csv"
prices = "../engage3project/prices.csv"
data = "../engage3project/data.csv"

#Read in CSV Data & Store into Pandas DF
auditors_df = pd.read_csv(auditors)
prices_df = pd.read_csv(prices)
data_df = pd.read_csv(data)

#Read & Open Json File
with open('stores.json') as f:
    store_data = json.load(f)

#Convert Json Dictionary to Pandas DataFrame
store_df = pd.DataFrame.from_dict(store_data)
```

```python
#Merge Prices, Auditors & Store dataframes into one to identify which Auditor is at fault
joined_df = pd.merge(prices_df, store_df, how='left', on=['Store ID', 'Store ID'])
combined_df = pd.merge(joined_df, auditors_df, how='left', on=['Auditor ID', 'Auditor ID'])
combined_df
```

```python
#Locate function to find who Auditor is for Whole Foods in the Kansas Regions
discrepancy_df = combined_df.loc[(combined_df["Banner"] == "Whole Foods") & (combined_df["Region_x"] == "Kansas")]
discrepancy_df
```

```python
discrepancy_df['Auditor ID'].unique()
#Auditor ID 713; Dave Johnson is our culprit
```
