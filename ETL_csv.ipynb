## Question 1 - ETL Data & Export to CSV

```python
#Import Dependencies

import pandas as pd
import numpy as np
import json
```

```python
#Load In Auditor & Price CSV's

auditors = "../engage3project/auditors.csv"
prices = "../engage3project/prices.csv"

#Read Data & Store into Pandas DF
auditors_df = pd.read_csv(auditors)
prices_df = pd.read_csv(prices)
```

```python
#Open & Load in Json File with Context Manager
with open('stores.json') as f:
    store_data = json.load(f)

#Convert Json Dictionary to Pandas DataFrame
store_df = pd.DataFrame.from_dict(store_data)
```

```python
#Question: Can you transform cross-tabulation of regional prices alongside each other, broken down by banner and written out into a spreadsheet?
#Create Banner DataFrame for Question 1
#Banner, UPC, Regional Prices

#Get 'Store ID' from store_df to add 'Region' data into prices_df
#Use Prices_Df to pull 'Price' data & 'UPC' to create new tables
```

```python
#Create joined_df merging store_df & prices_df
joined_df = pd.merge(prices_df, store_df, how='left', on=['Store ID', 'Store ID'])
joined_df
```

```python
joined_df.isna().sum()
```

```python
#Create clean df by removing unnecessary data in joined_df
banner_df = joined_df[['Banner', 'UPC','Price','Region']]
banner_df
```

```python
#Create pivot table creating 'Region' data into columns and using 'Prices' as values 
banner_df.pivot_table(values='Price', index=banner_df.index, columns='Region', aggfunc='first')
```

```python
#Convert pivot table to df without Reset Index
data = banner_df.pivot_table(index=['Banner','UPC'], columns='Region')
```

```python
#Convert MultiIndex to Columns
data.columns = ['Kansas', 'New York', 'Northern California', 'Texas']
```

```python
#Export finalized df to csv
data.to_csv('data.csv')
```
